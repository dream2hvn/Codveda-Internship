#  Data Cleaning and Preprocessing

Now, it’s time to clean and prepare the data for analysis. I first check for missing values ​​and handle them accordingly. If there are outliers, I identify and remove or adjust them to prevent them from skewing the analysis. I convert categorical variables to numeric representation using techniques such as label encoding. Finally, I may normalize or standardize the numeric data to ensure features are on the same scale.

```python
# Handle missing data
# Check for missing values
print(df.isnull().sum())
```

From this action there are no missing values, so no action is needed.


```python
# Detect and remove outliers
# Using the IQR method to detect outliers
# Select only numeric columns for IQR calculation
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Calculate Q1, Q3, and IQR for numeric columns
# Q1 represents the 25th percentile of the data
Q1 = df[numeric_columns].quantile(0.25)
# Q3 represents the 75th percentile of the data
Q3 = df[numeric_columns].quantile(0.75)
# IQR is calculated as the difference between Q3 and Q1
IQR = Q3 - Q1
```

```python
# Remove outliers
# This line filters the DataFrame 'df' to keep only rows that do not contain outliers.
df = df[~((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Convert categorical variables into numerical format
# Using Label Encoding for the 'species' column
label_encoder = LabelEncoder()
df['species'] = label_encoder.fit_transform(df['species'])
```

```python
# Normalize or standardize numerical data
# Using StandardScaler to standardize the numerical columns
scaler = StandardScaler()  # Create a StandardScaler object for feature scaling
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])  # Apply standardization to numerical features
```
